##### e. Analysis of Missing Values

```python
print(df.isna().sum())
```

```python
df_data = dict()
num_rows = int(df.shape[0])
for col in ['Age', 'Cabin', 'Embarked']:
  missing_values = int(df[col].isna().sum())
  df_data[col]= [num_rows-missing_values, missing_values]

print(f'Obtaining for each feature: [Number of Complete Data, Number of Missing Data] for the {df.shape[0]} entries:', end='\n\n')
print(df_data)
```

```python
import matplotlib.pyplot as plt

# colours
font_color = '#525252'
colors = ["darkturquoise", "lightcoral"]

# create a figure with one row and three columns for the graphs
fig, axes = plt.subplots(1, 3, figsize=(20, 10), facecolor='#e8f4f0')

# create a chart for each column
for i, col in enumerate(df_data):
  axes[i].pie(df_data[col], 
           labels=['Complete', 'Missing'], 
           startangle=90, 
           colors=colors, 
           textprops={'color':font_color, 'fontsize': 15},
           autopct='%1.1f%%')
  axes[i].set_title(col, fontsize=20, color=font_color)

title = fig.suptitle('Percentage of Missing Data by Feature', y=.9,
                     fontsize=30, color='black')

plt.show()
```

#### 2. Ages of the People Traveling on the Titanic <a class="anchor" id="ages"></a>

```python
# Calculate the mean, median, and mode
mean_age = df["Age"].mean()
median_age = df["Age"].median()
mode_age = df["Age"].mode()[0]  # mode() returns a Series, get the first value

# Plot the histogram
ax = df["Age"].hist(bins=15, color='teal', alpha=0.8, figsize=(20,10))
ax.set(xlabel='Age', ylabel='Count')

# Add vertical lines for mean, median, and mode
ax.axvline(mean_age, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_age:.2f}')
ax.axvline(median_age, color='green', linestyle='dashed', linewidth=2, label=f'Median: {median_age:.2f}')
ax.axvline(mode_age, color='blue', linestyle='dashed', linewidth=2, label=f'Mode: {mode_age:.2f}')

# Add a legend
ax.legend()

plt.show()
```

Usually, when the data distribution is not symmetrical, the most representative value is the median.

#### 3. Relationship Between Age and Survival/Death  <a class="anchor" id="age_survival"></a>

```python
import seaborn as sns

plt.figure(figsize=(20,10))
sns.kdeplot(df["Age"][df.Survived == 1], color="darkturquoise", fill=True)
sns.kdeplot(df["Age"][df.Survived == 0], color="lightcoral", fill=True)
plt.legend(['Survivors', 'Deceased'])
plt.title('Age Density Plot for the Surviving and Deceased Populations')
plt.grid()
plt.show()
```

We can observe that there is a peak in survival rates among children, which evidences the tendency to prioritize saving children. The classic phrase, "Women and children first," is fulfilled in the latter part.

#### 4. Relationship Between Gender and Survival/Death <a class="anchor" id="gender_survival"></a>

The number of people of each gender:

```python
df.groupby('Sex').count().rename(columns={'Name':'Count'})[['Count']]
```

We have 314 women and 577 men, meaning our data is not balanced. An imbalance in the data indicates that the number of women and men is not the same. This prevents us from being able to directly compare the survivors/deceased of each gender.

Therefore, to make a comparison, we will need to show percentages instead of the count.

```python
data_percent = df.groupby('Sex')['Survived'].value_counts(normalize=True).mul(100).rename('Percent').reset_index()
data_percent
```

```python
colors = ["darkturquoise", "lightcoral"] 
sns.catplot(x="Sex", y="Percent", hue="Survived", data=data_percent, kind="bar", palette=colors, height=10, aspect=0.8)
plt.show()
```

#### 5. Relationship Between Class and Survival/Death <a class="anchor" id="class_survival"></a>

Similarly to the previous case, we will analyze 'Pclass':

```python
df.groupby('Pclass').count().rename(columns={'Name':'Count'})[['Count']]
```

```python
data_percent = df.groupby('Pclass')['Survived'].value_counts(normalize=True).mul(100).rename('Percent').reset_index()
data_percent
```

```python
colors = ["darkturquoise", "lightcoral"] 
sns.catplot(x="Pclass", y="Percent", hue="Survived", data=data_percent, kind="bar", palette=colors, height=10, aspect=0.9)
plt.show();
```

We can see that the lower the class, the lower the survival rate.
